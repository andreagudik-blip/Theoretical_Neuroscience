# 1.1 trinogonometric functions

The sine and cosine functions describe periodic motion:

$$y_1​(t)=A_1​sin(ω_1​t)$$
$$y_2​(t)=A_2​cos(ω2​t)$$

Parameters

$A_1$, $A_2$. → amplitudes (maximum size of oscillation)

$\omega$ → angular frequency (rad/s)

t → time


Plot sin and cos, try changing the different parameters to see what happens to the functions.


```python

# Parameters
A1 = 1
A2 = 1
omega1 = 2*np.pi*1   # corresponds to f = 1 Hz
omega2 = 2*np.pi*1

# Time axis
t = np.linspace(0, 2, 1000)

# Define functions
y1 = A1 * np.sin(omega1 * t)
y2 = A2 * np.cos(omega2 * t)

# Plot
plt.plot(t, y1, label="y1 = A1 sin(ω1 t)")
plt.plot(t, y2, label="y2 = A2 cos(ω2 t)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.title("Sine and Cosine Oscillations")
plt.show()
```

## 1.1.3 Exponential function
The exponential function appears many places, especially in dynamical systems, since it can be a solution to a first
order differential equation. Plot the following exponential, which is commonly seen as exponential decay, e.g. in the
membrane potential after a synaptic input:

$ f(t)= e^{-t/\tau}\\$

Plot this with different values of τ. Try plotting on semi logarithmic scales (logarithmic y-scale).


 Large τ → slow decay


Small τ → fast decay

```python
# Time axis
t = np.linspace(0, 5, 500)

# Different time constants
taus = [0.5, 1, 2]

plt.figure()

for tau in taus:
    f = np.exp(-t/tau)
    plt.plot(t, f, label=f"τ = {tau}")

plt.xlabel("Time")
plt.ylabel("f(t)")
plt.title("Exponential decay")
plt.legend()
plt.show()
```

```python
#Semi log plot

plt.figure()

for tau in taus:
    f = np.exp(-t/tau)
    plt.semilogy(t, f, label=f"τ = {tau}")

plt.xlabel("Time")
plt.ylabel("f(t)")
plt.title("Exponential decay (semi-log plot)")
plt.legend()
plt.show()
```

power laws become straight in log-log plots.
Exponentials become straight in semi-log plots.

This is a diagnostic tool:

If data is straight in semi-log → exponential.
If straight in log-log → power law.

### 🔬 Physical Intuition (Membrane Example)

After a synaptic input:
$$ V(t)=V_0​e^{\frac{{−t}}{τ}}$$

τ tells you:


How fast the neuron forgets the input.


Large τ → slow decay → more temporal integration.


Small τ → fast decay → quick reset.

## Gaussian distribution
Defining a Gaussian distributions:
Parameters:

$\mu$= mean (center of the distribution)
$\sigma$ = standard deviation (width)

$\sigma^2$= variance

This is a probability density function (PDF).

It tells us how likely values near 𝑥 are.

Plot f (x) for different values of μ and σ . 


```python
def normal_pdf(x, mu, sigma):
    return (1/(sigma*np.sqrt(2*np.pi))) *  np.exp(-(x-mu)**2/(2*sigma**2))


x = np.linspace(-10, 10, 500) # 500 points between -10 and 10 

mu = 0
sigma = 1

y = normal_pdf(x, mu, sigma)

plt.plot(x, y)
plt.title("Normal distribution")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()           
```

## 1.1.6 Sigmoidal functions
Why is the sigmoidal function fundamental? It is found in medium with growth and saturation saturation. It is used
as smooth threshold/activation function, for simplified neuronal models. It can represent phase transitions, and
dose-response curves in pharmacology. It is also the most common neural network activation functions in artificial
networks. The simplest form look like this:

$\sigma(x)=\frac{1}{1+e^{-x}}$

Other functions, like hyperbolic tangent function, have sigmoidal shape. Choice of function depends on circum-
stances. Plot σ . Try putting a constant, a > 0 , in front of x in the function and plot it with different values of a,
starting with a = 1. Plot these in same coordinate system, for various values of a. What happens as a becomes
larger?

```python
def sigmoidal(x):
    return 1 / (1 + np.exp(-x))


x = np.linspace(-10, 10, 500) # 500 points between -10 and 10 



a_values= [-10,1,2,5,10,20,1000]

for a in a_values:
    plt.plot(x, sigmoidal(x * a), label=f"a={a}")

plt.title("sigmoid function for different values of a")
plt.xlabel("x")
plt.ylabel("sigma(x)")
plt.legend()
plt.show()
```

### 2.1.3 Finding the inverse operator

Assume two matrices are the following

```python
import sympy as sp

# Ukendte i invers-matricen
a, b, c, d = sp.symbols('a b c d')
Sinv = sp.Matrix([[a, b],
                  [c, d]])

I = sp.eye(2)

# Matricer fra opgaven
S1 = sp.Matrix([[1, 0],
                [0, 2]])

S2 = sp.Matrix([[1, 0],
                [0, 0]])

def solve_inverse_by_equations(S, name="S"):
    # Ligningerne fra S*Sinv = I
    eqs_matrix = sp.Eq(S * Sinv, I)
    # SymPy giver en matrix-ligning; gør den til 4 skalar-ligninger
    eqs = list((S * Sinv - I).reshape(4, 1))
    sol = sp.solve(eqs, [a, b, c, d], dict=True)

    print(f"\n{name} =")
    sp.pprint(S)

    if sol:
        sol = sol[0]
        Sinv_sol = sp.Matrix([[sol[a], sol[b]],
                              [sol[c], sol[d]]])
        print(f"\n{name}^-1 (løst fra ligninger) =")
        sp.pprint(Sinv_sol)

        print("\nTjek: S * S^-1 =")
        sp.pprint(S * Sinv_sol)
    else:
        print(f"\n{name} har ingen løsning til S*Sinv = I  → ingen inverse.")
        print("Årsag: determinanten er 0 (singulær matrix).")
        print("det(S) =", S.det())

# Kør for begge
solve_inverse_by_equations(S1, "S1")
solve_inverse_by_equations(S2, "S2")
```
