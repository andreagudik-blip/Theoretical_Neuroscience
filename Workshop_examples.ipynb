# 1.1 trinogonometric functions

The sine and cosine functions describe periodic motion:

$$y_1​(t)=A_1​sin(ω_1​t)$$
$$y_2​(t)=A_2​cos(ω2​t)$$

Parameters

$A_1$, $A_2$. → amplitudes (maximum size of oscillation)

$\omega$ → angular frequency (rad/s)

t → time


Plot sin and cos, try changing the different parameters to see what happens to the functions.


```python

# Parameters
A1 = 1
A2 = 1
omega1 = 2*np.pi*1   # corresponds to f = 1 Hz
omega2 = 2*np.pi*1

# Time axis
t = np.linspace(0, 2, 1000)

# Define functions
y1 = A1 * np.sin(omega1 * t)
y2 = A2 * np.cos(omega2 * t)

# Plot
plt.plot(t, y1, label="y1 = A1 sin(ω1 t)")
plt.plot(t, y2, label="y2 = A2 cos(ω2 t)")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.title("Sine and Cosine Oscillations")
plt.show()
```

## 1.1.3 Exponential function
The exponential function appears many places, especially in dynamical systems, since it can be a solution to a first
order differential equation. Plot the following exponential, which is commonly seen as exponential decay, e.g. in the
membrane potential after a synaptic input:

$ f(t)= e^{-t/\tau}\\$

Plot this with different values of τ. Try plotting on semi logarithmic scales (logarithmic y-scale).


 Large τ → slow decay


Small τ → fast decay

```python
# Time axis
t = np.linspace(0, 5, 500)

# Different time constants
taus = [0.5, 1, 2]

plt.figure()

for tau in taus:
    f = np.exp(-t/tau)
    plt.plot(t, f, label=f"τ = {tau}")

plt.xlabel("Time")
plt.ylabel("f(t)")
plt.title("Exponential decay")
plt.legend()
plt.show()
```

```python
#Semi log plot

plt.figure()

for tau in taus:
    f = np.exp(-t/tau)
    plt.semilogy(t, f, label=f"τ = {tau}")

plt.xlabel("Time")
plt.ylabel("f(t)")
plt.title("Exponential decay (semi-log plot)")
plt.legend()
plt.show()
```

power laws become straight in log-log plots.
Exponentials become straight in semi-log plots.

This is a diagnostic tool:

If data is straight in semi-log → exponential.
If straight in log-log → power law.

### 🔬 Physical Intuition (Membrane Example)

After a synaptic input:
$$ V(t)=V_0​e^{\frac{{−t}}{τ}}$$

τ tells you:


How fast the neuron forgets the input.


Large τ → slow decay → more temporal integration.


Small τ → fast decay → quick reset.

## Gaussian distribution
Defining a Gaussian distributions:
Parameters:

$\mu$= mean (center of the distribution)
$\sigma$ = standard deviation (width)

$\sigma^2$= variance

This is a probability density function (PDF).

It tells us how likely values near 𝑥 are.

Plot f (x) for different values of μ and σ . 


```python
def normal_pdf(x, mu, sigma):
    return (1/(sigma*np.sqrt(2*np.pi))) *  np.exp(-(x-mu)**2/(2*sigma**2))


x = np.linspace(-10, 10, 500) # 500 points between -10 and 10 

mu = 0
sigma = 1

y = normal_pdf(x, mu, sigma)

plt.plot(x, y)
plt.title("Normal distribution")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()           
```

Try plotting f (x) on a double-log plot for x > 0 and μ > 0. What happens to the nice bell-shape?

```python
x = np.linspace(0.1, 10, 500)   # avoid zero
mu = 3
sigma = 1

y = normal_pdf(x, mu, sigma)

plt.figure()
plt.loglog(x, y)
plt.title("Gaussian on log-log scale")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.show()
```
